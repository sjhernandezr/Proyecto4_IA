Limpieza de datos requerida para que el algoritmo funcione

```python
import pandas as pd
import numpy as np
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

df = pd.read_csv('../datos/Food_Delivery_Times.csv')

x = pd.get_dummies(df.drop(columns=['Delivery_Time_min', 'Order_ID'], errors='ignore'), dtype=float).replace([np.inf, -np.inf], np.nan).fillna(0)
y = df['Delivery_Time_min'].fillna(df['Delivery_Time_min'].median())


clf = DecisionTreeRegressor()
clf = clf.fit(x, y)

regresor = DecisionTreeRegressor(random_state=0)
scores = cross_val_score(regresor, x, y, cv=5)
print(-scores.mean())
```
